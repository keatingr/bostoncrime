```python
import csv
from StringIO import StringIO
from collections import namedtuple
```

```python
fields = header.replace(" ","_").replace("/","_").split(",")
print fields
```

```python
Crime = namedtuple('Crime', fields, verbose=True)
```

```python
def parse(row):
    reader = csv.reader(StringIO(row))
    row=reader.next()
    return Crime(*row)
```

```python
crimes = dataWoHeader.map(parse)
```

```python
crimes.first()
```

```python
crimes.first().INCIDENT_TYPE_DESCRIPTION
```

```python
crimes.map(lambda x:x.Year).countByValue()
```

```python
crimesFiltered=crimes.filter(lambda x:not(x.INCIDENT_TYPE_DESCRIPTION=="NA" or x.Year==''))\
        .filter(lambda x:int(x.Year)>=2006)
```

```python
crimesFiltered.map(lambda x:x.Year).countByValue()
```

```python
def extractCoords(location):
    location_lat = float(location[1:location.index(",")])
    location_lon = float(location[location.index(",")+1:-1])
    return (location_lat,location_lon)
```

```python
crimesFinal = crimesFiltered.filter(lambda x: extractCoords(x.Location)[0]!= 0 and \
                                   extractCoords(x.Location)[0]!=0 and \
                                    extractCoords(x.Location)[1]!=0 and \
                                   extractCoords(x.Location)[1]!=0)
```

```python
crimesFinal.first()
```

```python
# Trend by Year
crimesFinal.map(lambda x:x.Year).countByValue()
```

```python
crimesFinal.filter(lambda x:x.INCIDENT_TYPE_DESCRIPTION=='PubDrink')\
            .map(lambda x:x.Year)\
            .countByValue()
```

```python
import gmplot
gmap = gmplot.GoogleMapPlotter(42.1714138,-71.2938301, 16).from_geocode("Boston")
```

```python
b_lats = crimesFinal.filter(lambda x:x.INCIDENT_TYPE_DESCRIPTION=='PubDrink' and x.Year=='2015')\
    .map(lambda x:extractCoords(x.Location)[0])\
    .collect()
```

```python
b_lons = crimesFinal.filter(lambda x:x.INCIDENT_TYPE_DESCRIPTION=='PubDrink' and x.Year=='2015')\
    .map(lambda x:extractCoords(x.Location)[1])\
    .collect()
```

```python
gmap.scatter(b_lats,b_lons,'#4440C8',size=40,marker=False)
```
